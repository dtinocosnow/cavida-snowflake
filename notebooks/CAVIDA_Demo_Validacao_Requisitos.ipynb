```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from IPython.display import display, HTML, Markdown
import warnings
warnings.filterwarnings('ignore')

REQUISITOS = {
    3:  {'cat':'Arquitetura',   'pri':'Must',   'nome':'Separação Compute/Storage e Isolamento',   'resp':''},
    5:  {'cat':'Arquitetura',   'pri':'Should', 'nome':'Controlo de Versões',                       'resp':''},
    6:  {'cat':'Arquitetura',   'pri':'Should', 'nome':'Time Travel',                                 'resp':''},
    8:  {'cat':'Integração',    'pri':'Must',   'nome':'Integração RDBMS',                           'resp':''},
    9:  {'cat':'Integração',    'pri':'Must',   'nome':'Dados Não Estruturados, APIs, SFTP',           'resp':''},
    10: {'cat':'Integração',    'pri':'Should', 'nome':'Aplicações Web Governadas',                   'resp':''},
    13: {'cat':'Governança',    'pri':'Must',   'nome':'Catálogo Centralizado',                       'resp':''},
    14: {'cat':'Governança',    'pri':'Should', 'nome':'Linhagem Técnica',                            'resp':''},
    16: {'cat':'Governança',    'pri':'Could',  'nome':'Qualidade de Dados',                          'resp':''},
    17: {'cat':'Segurança',     'pri':'Must',   'nome':'Residência UE',                               'resp':''},
    21: {'cat':'Segurança',     'pri':'Could',  'nome':'Masking Dinâmico e RLS',                      'resp':''},
    30: {'cat':'Operações',    'pri':'Should', 'nome':'Orquestração',                                'resp':''},
    32: {'cat':'Operações',    'pri':'Should', 'nome':'Ferramentas de Pipelines',                    'resp':''},
    34: {'cat':'FinOps',        'pri':'Must',   'nome':'Controlo de Custos',                          'resp':''},
    35: {'cat':'FinOps',        'pri':'Should', 'nome':'Atribuição de Custos',                        'resp':''},
    36: {'cat':'FinOps',        'pri':'Should', 'nome':'Otimização de Custos',                        'resp':''},
    37: {'cat':'BI',            'pri':'Should', 'nome':'Camada Semântica',                             'resp':''},
    42: {'cat':'IA/GenAI',      'pri':'Could',  'nome':'Capacidades IA/GenAI',                        'resp':''},
    46: {'cat':'NLQ',           'pri':'Could',  'nome':'Linguagem Natural',                           'resp':''},
    47: {'cat':'Referências',   'pri':'Should', 'nome':'Referências Seguros',                         'resp':''},
}

checkpoints = {}

def checkpoint(req_id, resultado, descricao):
    ok = resultado is not None and (isinstance(resultado, bool) and resultado or isinstance(resultado, (int, float)) and resultado > 0 or isinstance(resultado, pd.DataFrame) and len(resultado) > 0 or isinstance(resultado, str) and len(resultado) > 0)
    checkpoints[req_id] = {'ok': ok, 'desc': descricao}
    REQUISITOS[req_id]['resp'] = descricao
    icone = '✅' if ok else '❌'
    cor = '#2ecc71' if ok else '#e74c3c'
    pri = REQUISITOS[req_id]['pri']
    badge_cor = {'Must':'#e74c3c','Should':'#f39c12','Could':'#3498db'}[pri]
    html = f"""
    <div style='background:linear-gradient(135deg,#1a1a2e,#16213e);border-left:5px solid {cor};
        padding:15px 20px;margin:10px 0;border-radius:8px;font-family:system-ui;'>
        <div style='display:flex;align-items:center;gap:12px;'>
            <span style='font-size:28px;'>{icone}</span>
            <div>
                <span style='background:{badge_cor};color:white;padding:2px 8px;border-radius:4px;
                    font-size:11px;font-weight:bold;'>{pri}</span>
                <span style='color:white;font-size:16px;font-weight:bold;margin-left:8px;'>
                    REQ {req_id} — {REQUISITOS[req_id]['nome']}</span>
            </div>
        </div>
        <div style='color:#b0b0b0;margin-top:8px;padding-left:42px;font-size:13px;'>
            {descricao}
        </div>
    </div>
    """
    display(HTML(html))

def mostrar_progresso():
    total = len(REQUISITOS)
    feitos = len(checkpoints)
    ok = sum(1 for v in checkpoints.values() if v['ok'])
    pct = feitos / total * 100
    bar_ok = ok / total * 100
    bar_fail = (feitos - ok) / total * 100
    html = f"""
    <div style='background:#1a1a2e;padding:15px 20px;border-radius:8px;margin:10px 0;font-family:system-ui;'>
        <div style='color:white;font-size:14px;margin-bottom:8px;'>
            📊 <b>Progresso:</b> {feitos}/{total} requisitos validados ·
            <span style='color:#2ecc71;'>✅ {ok} OK</span> ·
            <span style='color:#e74c3c;'>❌ {feitos-ok} Falha</span> ·
            <span style='color:#95a5a6;'>⏳ {total-feitos} Pendente</span>
        </div>
        <div style='background:#2c3e50;border-radius:6px;height:24px;overflow:hidden;display:flex;'>
            <div style='width:{bar_ok}%;background:linear-gradient(90deg,#27ae60,#2ecc71);height:100%;'></div>
            <div style='width:{bar_fail}%;background:linear-gradient(90deg,#c0392b,#e74c3c);height:100%;'></div>
        </div>
    </div>
    """
    display(HTML(html))

display(HTML("<div style='background:linear-gradient(135deg,#0f3460,#16213e);padding:20px;border-radius:10px;text-align:center;'><span style='color:#2ecc71;font-size:24px;'>⚙️ Sistema de Checkpoints Inicializado</span><br><span style='color:#b0b0b0;'>20 requisitos para validar</span></div>"))
```

## REQ 5 — Controlo de Versões (Should)

| Capacidade | Detalhe Snowflake |
|---|---|
| **Git nativo** | Repositórios ligados ao Snowflake |
| **CREATE OR REPLACE** | Versionamento de stored procedures |
| **UNDROP** | Recuperação de objetos eliminados |

```python
%%sql -r req5_sp
CREATE OR REPLACE PROCEDURE CAVIDA_DEMO.DEV.SP_CLAIMS_SUMMARY()
  RETURNS VARCHAR LANGUAGE SQL EXECUTE AS CALLER AS
$$
BEGIN
  CREATE OR REPLACE TABLE CAVIDA_DEMO.DEV.CLAIMS_SUMMARY AS
  SELECT DATE_TRUNC('month',c.CLAIM_DATE) AS MES, c.CLAIM_TYPE AS TIPO,
    COUNT(*) AS N, SUM(c.PAID_AMOUNT) AS TOTAL
  FROM CAVIDA_DEMO.PROD.FACT_CLAIMS c GROUP BY 1,2;
  RETURN 'OK: '||(SELECT COUNT(*) FROM CAVIDA_DEMO.DEV.CLAIMS_SUMMARY)||' linhas';
END;
$$;
CALL CAVIDA_DEMO.DEV.SP_CLAIMS_SUMMARY()
```

```python
sp_ok = req5_sp is not None and len(req5_sp) > 0
resultado_sp = str(req5_sp.iloc[0,0]) if sp_ok else 'Falhou'
checkpoint(5, sp_ok,
    f"Stored Procedure criada com CREATE OR REPLACE · Resultado: {resultado_sp} · Git + UNDROP disponíveis")
mostrar_progresso()
```

## REQ 16 — Qualidade de Dados (Could)

| Capacidade | Detalhe |
|---|---|
| **DMFs** | NULL_COUNT, DUPLICATE_COUNT, FRESHNESS |
| **Cortex Data Quality** | Sugestões IA automáticas |
| **Alertas** | Ações automáticas por threshold |

```python
%%sql -r req16_dq
SELECT 'Sinistros órfãos' AS VERIFICACAO,
  (SELECT COUNT(*) FROM CAVIDA_DEMO.PROD.FACT_CLAIMS c
   LEFT JOIN CAVIDA_DEMO.PROD.FACT_POLICIES p ON c.POLICY_ID=p.POLICY_ID
   WHERE p.POLICY_ID IS NULL) AS RESULTADO,
  CASE WHEN (SELECT COUNT(*) FROM CAVIDA_DEMO.PROD.FACT_CLAIMS c
   LEFT JOIN CAVIDA_DEMO.PROD.FACT_POLICIES p ON c.POLICY_ID=p.POLICY_ID
   WHERE p.POLICY_ID IS NULL) = 0 THEN '✅ OK' ELSE '⚠️ Atenção' END AS STATUS
UNION ALL
SELECT 'Prémios negativos',
  (SELECT COUNT(*) FROM CAVIDA_DEMO.PROD.FACT_POLICIES WHERE PREMIUM_AMOUNT<0),
  CASE WHEN (SELECT COUNT(*) FROM CAVIDA_DEMO.PROD.FACT_POLICIES WHERE PREMIUM_AMOUNT<0) = 0 THEN '✅ OK' ELSE '⚠️ Atenção' END
UNION ALL
SELECT 'Duplicados POLICY_NUMBER',
  (SELECT COUNT(*)-COUNT(DISTINCT POLICY_NUMBER) FROM CAVIDA_DEMO.PROD.FACT_POLICIES),
  CASE WHEN (SELECT COUNT(*)-COUNT(DISTINCT POLICY_NUMBER) FROM CAVIDA_DEMO.PROD.FACT_POLICIES) = 0 THEN '✅ OK' ELSE '⚠️ Atenção' END
```

```python
dq_ok = req16_dq is not None and len(req16_dq) > 0
n_checks = len(req16_dq) if dq_ok else 0
n_ok = len(req16_dq[req16_dq['STATUS'].str.contains('OK')]) if dq_ok else 0

if dq_ok:
    display(req16_dq.style.set_properties(**{'text-align':'center'}).set_caption('🔍 Verificações de Qualidade'))

checkpoint(16, dq_ok,
    f"{n_checks} verificações executadas · {n_ok}/{n_checks} OK · DMFs + Alertas disponíveis")
mostrar_progresso()
```
